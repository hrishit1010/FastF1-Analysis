```python
import pandas as pd

# List of all your CSV filenames
files = [
    'australia_practice_times_2024.csv',
    'austria_practice_times_2024.csv',
    'bahrain_practice_times_2024.csv',
    'belgium_practice_times_2024.csv',
    'britain_practice_times_2024.csv',
    'china_practice_times_2024.csv',
    'dutch_practice_times_2024.csv',
    'emilia_romagna_practice_times_2024.csv',
    'hungary_practice_times_2024.csv',
    'italy_practice_times_2024.csv',
    'japan_practice_times_2024.csv',
    'miami_practice_times_2024.csv',
    'monaco_practice_times_2024.csv',
    'saudi_arabia_practice_times_2024.csv',
    'spain_practice_times_2024.csv'
]

# Read each CSV into a DataFrame and store them in a list
dfs = [pd.read_csv(file) for file in files]

# Concatenate all DataFrames into one
combined_df = pd.concat(dfs, ignore_index=True)

# Check the combined dataframe
print(combined_df.shape)
print(combined_df.head())
```

```python
#fill missing values with mean
time_cols = ['LapTime', 'PitOutTime', 'PitInTime', 'Sector1Time', 'Sector2Time', 'Sector3Time']

for col in time_cols:
    if col in combined_df.columns:
        # Convert to timedelta
        combined_df[col] = pd.to_timedelta(combined_df[col])
        # Converting total seconds as float
        seconds = combined_df[col].dt.total_seconds()
        # Mean
        mean_seconds = seconds.mean()
        seconds_filled = seconds.fillna(mean_seconds)
        # Convert back to timedelta
        combined_df[col] = pd.to_timedelta(seconds_filled, unit='s')
```

```python
columns_to_fill = [
    'Sector1SessionTime', 'Sector2SessionTime', 'Sector3SessionTime',
    'SpeedI1', 'SpeedI2', 'SpeedFL', 'SpeedST',
    'IsPersonalBest', 'LapStartDate', 'Position', 'DeletedReason'
]

for col in columns_to_fill:
    if col in combined_df.columns:
        # Convert column to numeric, coercing errors to NaN
        combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')

for col in columns_to_fill:
    if col in combined_df.columns:

        mean_value = combined_df[col].mean(skipna=True)
        # Fill NaN with mean value
        combined_df[col].fillna(mean_value, inplace=True)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Converting LapTime from string to timedelta
Data['LapTime'] = pd.to_timedelta(Data['LapTime'])


Data['LapTime_seconds'] = Data['LapTime'].dt.total_seconds()


print(Data[['LapTime', 'LapTime_seconds']].head())

#LapTime trend over laps by driver
plt.figure(figsize=(15, 8))
sns.lineplot(x='LapNumber', y='LapTime_seconds', hue='Driver', data=Data, marker='o')
plt.title('Lap Time Trend Over Laps by Driver')
plt.xlabel('Lap Number')
plt.ylabel('Lap Time (seconds)')
plt.show()

```

This line plot illustrates how each driver's lap time evolves from the beginning of the race through to the end (lap by lap).
The x-axis indicates the lap number, and the y-axis the lap time in seconds.
Various colored lines indicate various drivers, making it easy to directly compare trends in performance.
Seeing this plot can assist in identifying:
- Consistency: Drivers whose lap times are consistent show consistent performance

- Improvement: Drivers whose lap times become better across laps might be warming up or optimizing their speed.
- Degradation: Drivers whose lap times are increasing might be experiencing tire degradation, fatigue, or changing track conditions.
- Outliers or incidents: Sudden, unexplained jumps in lap time may be caused by mistakes, pit stops, or engine problems.

This plot is useful for a coach, spectator, or analyst to look at driver pacing strategy and endurance during a race.

```python
import fastf1
from fastf1 import plotting
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Caching Done to speed up repeated queries
fastf1.Cache.enable_cache('fastf1_cache')

#Races in cities
races = [
    'australia', 'austria', 'bahrain', 'belgium', 'britain', 'china', 'dutch',
    'emilia_romagna', 'hungary', 'italy', 'japan', 'miami', 'monaco',
    'saudi_arabia', 'spain'
]

year = 2023

all_laps = []

for race in races:
    try:
        #qualifying session laps; Q= Qualifying
        session = fastf1.get_session(year, race, 'Q')
        session.load()

        laps = session.laps

        laps_df = laps[['Driver', 'LapTime']].copy()
        laps_df['City'] = race

        all_laps.append(laps_df)
        print(f'Loaded {race}')
    except Exception as e:
        print(f'Failed to load {race}: {e}')

# Combining all laps into one DataFrame
df_all_laps = pd.concat(all_laps)

# Remove laps with missing LapTime
df_all_laps = df_all_laps[df_all_laps['LapTime'].notna()]

# Converting LapTime (timedelta) to seconds
df_all_laps['LapTime_seconds'] = df_all_laps['LapTime'].dt.total_seconds()

# Calculating best average lap time per Driver per City
avg_lap_time = df_all_laps.groupby(['Driver', 'City'])['LapTime_seconds'].mean().reset_index()

# Finding the city with best average lap time per driver
best_city_per_driver = avg_lap_time.loc[avg_lap_time.groupby('Driver')['LapTime_seconds'].idxmin()]

print(best_city_per_driver)

# Plot best average lap time city per driver
plt.figure(figsize=(16, 8))
sns.barplot(x='Driver', y='LapTime_seconds', hue='City', data=best_city_per_driver, dodge=False)
plt.title(f'Best Average Lap Time City per Driver - {year} Qualifying Sessions')
plt.ylabel('Average Lap Time (seconds)')
plt.xlabel('Driver')
plt.legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
